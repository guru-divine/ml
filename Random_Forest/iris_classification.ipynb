```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.data.shape
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue')
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(criterion='entropy', n_estimators=200)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
scores = []
n_estimators = []
for i in range(1, 101, 5):
    models = RandomForestClassifier(n_estimators=i)
    models.fit(X_train, y_train)
    score = models.score(X_test, y_test)
    scores.append(score)
    n_estimators.append(i)
    print(f'n_estimators = {i}, score = {score}')
plt.scatter(n_estimators, scores)
plt.xlabel('n_estimators')
plt.ylabel('score')
```
