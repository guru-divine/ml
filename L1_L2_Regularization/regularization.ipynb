```python
import pandas as pd
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize',
               'BuildingArea', 'CouncilArea', 'Propertycount', 'Price']
dataset = df[cols_to_use]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Bedroom2', 'Bathroom', 'Car', 'Propertycount']
cols_to_fill_mean = ['Landsize', 'BuildingArea']

dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset[cols_to_fill_mean] = dataset[cols_to_fill_mean].fillna(dataset[cols_to_fill_mean].mean())
```

```python
dataset.describe()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
cols_to_convert_to_binary = ['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea']
for col in cols_to_convert_to_binary:
    dataset[col] = le.fit_transform(dataset[col])
```

```python
dataset.head()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
X = dataset.drop(['Price'], axis='columns')
y = dataset.Price
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)

lasso_regression_model = Lasso(alpha=50, max_iter=200, tol=0.1).fit(X_train, y_train)

ridge_regression_model = Ridge(alpha=50, max_iter=100, tol=0.1).fit(X_train, y_train)
```

```python
train_results = []
test_results = []
models = [linear_regression_model, lasso_regression_model, ridge_regression_model]

for model in models:
    train_results.append(model.score(X_train, y_train))
    test_results.append(model.score(X_test, y_test))
```

```python
result_df = pd.DataFrame({
    'models': models,
    'train_results': train_results,
    'test_results': test_results
})
result_df
```
