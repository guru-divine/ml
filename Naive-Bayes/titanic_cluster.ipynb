```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
df.head()
```

```python
df.describe()
```

```python
import numpy as np
```

```python
def cnt_survived(val, col, df):
    cnt = np.zeros(val+1)
    for index, row in df.iterrows():
        if row['Survived'] == 1:
            cnt[int(row[col])] += 1
    return cnt
```

```python
print("Pclass Cnt survived: ", cnt_survived(3, 'Pclass', df))
print("SibSp Cnt survived", cnt_survived(8, 'SibSp', df))
print("Parch Cnt survived", cnt_survived(8, 'Parch', df))
```

```python
#PClass is of no use
df = df.drop(['Pclass'], axis='columns')
df.head()
```

```python
df['Sex'] = df['Sex'].apply(lambda x: 1 if x=='male' else 0)
df.head()
```

```python
df.isna().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
X = df.drop(['Survived'], axis='columns')
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
nb.score(X_test, y_test)
```
